```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_5871594.csv')
metadata1= pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_1098332.csv')
metadata2= pd.read_csv('Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_1098332.csv')
```

```python
metadata1.info()
```

```python
merged_df = pd.merge(df, metadata1, on= 'Country Code' , how='left')
```

```python
merged_df.info()
```

```python
pd.set_option('display.max_rows',None)
merged_df["Country Name"]
```

```python
filtered_data = merged_df.tail(50)
#creating the bar chart
color = ("blue")
fig = plt.figure(figsize =(20,4))
plt.bar (filtered_data['Country Name'], filtered_data['1960'],color=color)
plt.xlabel ('Country Name')
plt.ylabel ('Population in 1960')
plt.title('Population of selected countries in 1960')
plt.xticks(rotation=45, ha='right')
plt.show()
```
